```python
# =============================================================================
# SECTION 1: MACHINE LEARNING MODEL TRAINING & EVALUATION
# =============================================================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import pickle
import os

from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score, RandomizedSearchCV, learning_curve, validation_curve
from sklearn.preprocessing import LabelEncoder, StandardScaler, MinMaxScaler
from sklearn.metrics import (classification_report, confusion_matrix, accuracy_score,
                             roc_curve, auc, precision_recall_curve)
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, StackingClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.utils.class_weight import compute_class_weight

from imblearn.over_sampling import SMOTE, SMOTENC, ADASYN
from imblearn.under_sampling import TomekLinks
from imblearn.combine import SMOTETomek

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau

warnings.filterwarnings("ignore")

# =============================================================================
# DATA LOADING & INITIAL EXPLORATION
# =============================================================================
def load_dataset(filepath):
    """
    Load dataset from an Excel file and perform initial inspection.
    """
    try:
        df = pd.read_excel(filepath)
        print("Dataset loaded successfully!")
        display(df.head())
        display(df.dtypes)
        display(df.describe())
        display(df['MSD_Risk_Level'].value_counts())
    except FileNotFoundError:
        print(f"Error: '{filepath}' not found.")
        return None
    except KeyError:
        print("Error: 'MSD_Risk_Level' column not found in dataset.")
        return None
    return df

def exploratory_data_analysis(df):
    """
    Perform EDA: missing values, distributions, outlier detection, correlations, and pairplots.
    """
    # Missing values
    missing_values = df.isnull().sum()
    missing_percent = (missing_values / len(df)) * 100
    print("Missing Values:\n", missing_values)
    print("\nMissing Value Percentage:\n", missing_percent)

    # Distributions and boxplots for numerical features
    numerical_features = df.select_dtypes(include=['number']).columns
    for col in numerical_features:
        plt.figure(figsize=(8, 6))
        sns.histplot(df[col], kde=True)
        plt.title(f'Distribution of {col}')
        plt.show()

        plt.figure(figsize=(8, 6))
        sns.boxplot(x=df[col])
        plt.title(f'Boxplot of {col}')
        plt.show()

    # Correlation matrix
    plt.figure(figsize=(12, 10))
    sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True, cmap='coolwarm')
    plt.title('Correlation Matrix (Numerical Features)')
    plt.show()

    # Pairplot for selected features
    try:
        selected_features = ['Age', 'Height', 'Weight', 'Pain_Score', 'Backpack_Weight', 'MSD_Risk_Level']
        sns.pairplot(df[selected_features], hue='MSD_Risk_Level')
        plt.show()
    except Exception as e:
        print(f"Pairplot error: {e}")
        try:
            sns.pairplot(df.select_dtypes(include=['number']), hue='MSD_Risk_Level')
            plt.show()
        except Exception as e:
            print(f"Pairplot error (fallback): {e}")

# =============================================================================
# PREPROCESSING & FEATURE ENGINEERING
# =============================================================================
def preprocess_data(df):
    """
    Preprocess the data by handling outliers, encoding, scaling, oversampling,
    and creating interaction/polynomial features.

    This function now also encodes the target variable 'MSD_Risk_Level'.

    Returns:
        df_resampled: The preprocessed DataFrame.
        le_target: The fitted LabelEncoder for the target variable.
    """
    # One-hot encode Backpack_Position if present
    if 'Backpack_Position' in df.columns:
        df = pd.get_dummies(df, columns=['Backpack_Position'], prefix='Backpack_Pos')

    # Winsorize numerical columns (exclude target)
    numerical_cols = df.select_dtypes(include=np.number).columns
    for col in numerical_cols:
        if col != 'MSD_Risk_Level':
            q1 = df[col].quantile(0.25)
            q3 = df[col].quantile(0.75)
            iqr = q3 - q1
            lower_bound = q1 - 1.5 * iqr
            upper_bound = q3 + 1.5 * iqr
            df[col] = np.clip(df[col], lower_bound, upper_bound)

    # Label encode Gender if exists
    if 'Gender' in df.columns:
        le_gender = LabelEncoder()
        df['Gender'] = le_gender.fit_transform(df['Gender'])

    # --- Oversample using SMOTE ---
    X = df.drop('MSD_Risk_Level', axis=1)
    y = df['MSD_Risk_Level']
    smote = SMOTE(random_state=42)
    X_resampled, y_resampled = smote.fit_resample(X, y)
    df_resampled = pd.DataFrame(X_resampled, columns=X.columns)
    df_resampled['MSD_Risk_Level'] = y_resampled

    # Log-transform skewed features (excluding target and Gender)
    numerical_cols = df_resampled.select_dtypes(include=np.number).columns
    for col in numerical_cols:
        if col not in ['MSD_Risk_Level', 'Gender']:
            if df_resampled[col].min() <= 0:
                df_resampled[col] = df_resampled[col] - df_resampled[col].min() + 1
            df_resampled[col] = np.log1p(df_resampled[col])

    # Scale numerical features using MinMaxScaler (excluding target)
    features_to_scale = numerical_cols.drop('MSD_Risk_Level', errors='ignore')
    scaler = MinMaxScaler()
    df_resampled[features_to_scale] = scaler.fit_transform(df_resampled[features_to_scale])

    # Create interaction and polynomial features if key columns are present
    required_cols = ['Weight', 'Backpack_Weight', 'Age', 'Pain_Score']
    if all(col in df_resampled.columns for col in required_cols):
        df_resampled['Weight_BackpackWeight'] = df_resampled['Weight'] * df_resampled['Backpack_Weight']
        df_resampled['Age_PainScore'] = df_resampled['Age'] * df_resampled['Pain_Score']
        df_resampled['Pain_Score_Squared'] = df_resampled['Pain_Score'] ** 2
        df_resampled['Backpack_Weight_Cubed'] = df_resampled['Backpack_Weight'] ** 3
        df_resampled['Age_Squared'] = df_resampled['Age'] ** 2

    # --------------------
    # Encode the target variable so that it is numeric
    le_target = LabelEncoder()
    df_resampled['MSD_Risk_Level'] = le_target.fit_transform(df_resampled['MSD_Risk_Level'])

    print("Preprocessed Data:")
    display(df_resampled.head())
    return df_resampled, le_target

def split_data(df):
    """
    Split the data into training, validation, and test sets using stratification.
    """
    X = df.drop('MSD_Risk_Level', axis=1)
    y = df['MSD_Risk_Level']

    # Train+validation (80%) and test (20%)
    X_train_val, X_test, y_train_val, y_test = train_test_split(
        X, y, test_size=0.2, stratify=y, random_state=42
    )
    # Train (75% of train_val) and validation (25% of train_val)
    X_train, X_val, y_train, y_val = train_test_split(
        X_train_val, y_train_val, test_size=0.25, stratify=y_train_val, random_state=42
    )

    print("Data Shapes:")
    print("X_train:", X_train.shape, "y_train:", y_train.shape)
    print("X_val:", X_val.shape, "y_val:", y_val.shape)
    print("X_test:", X_test.shape, "y_test:", y_test.shape)
    print("\nTraining set class distribution:\n", y_train.value_counts())
    return X_train, X_val, X_test, y_train, y_val, y_test

# =============================================================================
# MODEL BUILDING & TRAINING
# =============================================================================
def build_stacking_ensemble():
    """
    Build a stacking ensemble with multiple base classifiers and a logistic regression meta-model.
    """
    base_models = [
        ('rf', RandomForestClassifier(n_estimators=200, max_depth=None, class_weight='balanced', random_state=42)),
        ('gb', GradientBoostingClassifier(n_estimators=200, random_state=42)),
        # Removed deprecated 'use_label_encoder' parameter from XGBClassifier
        ('xgb', XGBClassifier(eval_metric='mlogloss', random_state=42)),
        ('lgbm', LGBMClassifier(class_weight='balanced', random_state=42)),
        ('svc', SVC(probability=True, class_weight='balanced', random_state=42))
    ]
    # Added class_weight='balanced' to meta-model
    meta_model = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=1000, class_weight='balanced')
    stacking_clf = StackingClassifier(
        estimators=base_models,
        final_estimator=meta_model,
        cv=3,  # Consistent with hyperparameter tuning CV
        stack_method='predict_proba',
        n_jobs=-1
    )
    return stacking_clf

def train_and_evaluate_models(X_train, X_val, X_test, y_train, y_val, y_test):
    """
    Train both the stacking ensemble and neural network models.
    Evaluate their performance on the validation and test sets.
    """
    results = {}
    histories = {}

    # --- Stacking Ensemble ---
    ensemble_model = build_stacking_ensemble()
    ensemble_model.fit(X_train, y_train)

    # Evaluate on validation set instead of using cross_val_score on training data
    val_preds = ensemble_model.predict(X_val)
    val_acc = accuracy_score(y_val, val_preds)
    test_preds = ensemble_model.predict(X_test)
    test_acc = accuracy_score(y_test, test_preds)

    results['ensemble'] = {
        'model': ensemble_model,
        'val_accuracy': val_acc,
        'val_report': classification_report(y_val, val_preds, output_dict=True),
        'test_accuracy': test_acc,
        'test_report': classification_report(y_test, test_preds, output_dict=True),
        'test_preds': test_preds
    }

    # Print ensemble model evaluation results
    print(f"\nEnsemble Model Validation Accuracy: {val_acc:.4f}")
    print("Ensemble Model Validation Classification Report:")
    print(classification_report(y_val, val_preds))



    return results

# =============================================================================
# HYPERPARAMETER TUNING FOR ENSEMBLE (STACKING) MODEL
# =============================================================================

def hyperparameter_tuning_ensemble_fast(X_train, y_train, X_val, y_val, param_grid, n_iter=5, cv=3):
    """
    Perform hyperparameter tuning for the stacking ensemble using RandomizedSearchCV
    with a reduced parameter grid and fewer iterations for faster tuning.

    Parameters:
        X_train (DataFrame): Training features.
        y_train (Series): Training labels.
        X_val (DataFrame): Validation features.
        y_val (Series): Validation labels.
        param_grid (dict): Reduced parameter grid.
        n_iter (int): Fewer iterations (e.g., 5).
        cv (int): Number of cross-validation folds (e.g., 3).

    Returns:
        best_estimator: The best stacking ensemble model.
        best_params (dict): Best hyperparameters found.
    """
    ensemble_model = build_stacking_ensemble()

    # Create a folder for models if it doesn't exist
    models_dir = "models"
    if not os.path.exists(models_dir):
        os.makedirs(models_dir)

    save_path = os.path.join(models_dir, "best_ensemble_model_fast.pkl")

    # Use RandomizedSearchCV with reduced grid and iterations
    random_search = RandomizedSearchCV(
        estimator=ensemble_model,
        param_distributions=param_grid,
        n_iter=n_iter,           # Reduced number of iterations
        scoring='accuracy',
        cv=cv,                   # Use fewer folds if needed
        random_state=42,
        n_jobs=-1,
        verbose=2
    )

    # Optionally, combine training and validation sets for tuning:
    X_combined = pd.concat([X_train, X_val])
    y_combined = pd.concat([y_train, y_val])

    random_search.fit(X_combined, y_combined)
    best_estimator = random_search.best_estimator_
    best_params = random_search.best_params_
    print("Best Ensemble Hyperparameters (Fast Tuning):", best_params)
    print("Best Cross-Validated Accuracy (Fast Tuning):", random_search.best_score_)

    # Save the best ensemble model to a pickle file
    try:
        with open(save_path, "wb") as f:
            pickle.dump(best_estimator, f)
        print(f"Best ensemble model (fast tuning) saved to {save_path}")
    except Exception as e:
        print(f"Error saving model: {e}")

    return best_estimator, best_params
# Updated parameter grid
def get_ensemble_param_grid_fast():
    """
    Returns a reduced parameter grid for ensemble hyperparameter tuning.
    This grid is smaller than the full one to speed up tuning.
    """
    param_grid = {
        'rf__n_estimators': [100, 200],
        'rf__max_depth': [None, 10],
        'gb__n_estimators': [100, 200],
        'xgb__max_depth': [3, 5],
        'lgbm__n_estimators': [100, 200],
        'svc__C': [0.1, 1],
        'final_estimator__C': [0.1, 1]
    }
    return param_grid

# =============================================================================
# IMPROVED MODEL EVALUATION & VISUALIZATION
# =============================================================================
def evaluate_ensemble_model(model, X_train, X_val, X_test, y_train, y_val, y_test):
    """
    Comprehensive evaluation of the ensemble model with visualization.
    """
    # Make predictions
    train_preds = model.predict(X_train)
    val_preds = model.predict(X_val)
    test_preds = model.predict(X_test)

    # Get prediction probabilities for ROC curves
    train_probs = model.predict_proba(X_train)
    val_probs = model.predict_proba(X_val)
    test_probs = model.predict_proba(X_test)

    # Get accuracy scores
    train_acc = accuracy_score(y_train, train_preds)
    val_acc = accuracy_score(y_val, val_preds)
    test_acc = accuracy_score(y_test, test_preds)

    print(f"Ensemble Model Performance:")
    print(f"Training Accuracy: {train_acc:.4f}")
    print(f"Validation Accuracy: {val_acc:.4f}")
    print(f"Test Accuracy: {test_acc:.4f}")

    # Classification reports
    print("\nTraining Classification Report:")
    print(classification_report(y_train, train_preds))
    print("\nValidation Classification Report:")
    print(classification_report(y_val, val_preds))
    print("\nTest Classification Report:")
    print(classification_report(y_test, test_preds))

    # Confusion matrices
    plt.figure(figsize=(18, 6))

    plt.subplot(1, 3, 1)
    train_cm = confusion_matrix(y_train, train_preds)
    sns.heatmap(train_cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(f'Training Confusion Matrix\nAccuracy: {train_acc:.4f}')
    plt.xlabel('Predicted')
    plt.ylabel('True')

    plt.subplot(1, 3, 2)
    val_cm = confusion_matrix(y_val, val_preds)
    sns.heatmap(val_cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(f'Validation Confusion Matrix\nAccuracy: {val_acc:.4f}')
    plt.xlabel('Predicted')
    plt.ylabel('True')

    plt.subplot(1, 3, 3)
    test_cm = confusion_matrix(y_test, test_preds)
    sns.heatmap(test_cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(f'Test Confusion Matrix\nAccuracy: {test_acc:.4f}')
    plt.xlabel('Predicted')
    plt.ylabel('True')

    plt.tight_layout()
    plt.show()

    # ROC curves for multiclass classification (one-vs-rest)
    n_classes = len(np.unique(y_train))

    plt.figure(figsize=(12, 8))
    for i in range(n_classes):
        # Calculate ROC curve and ROC area for each class
        fpr_train, tpr_train, _ = roc_curve(y_train == i, train_probs[:, i])
        fpr_val, tpr_val, _ = roc_curve(y_val == i, val_probs[:, i])
        fpr_test, tpr_test, _ = roc_curve(y_test == i, test_probs[:, i])

        roc_auc_train = auc(fpr_train, tpr_train)
        roc_auc_val = auc(fpr_val, tpr_val)
        roc_auc_test = auc(fpr_test, tpr_test)

        plt.plot(fpr_train, tpr_train, linestyle=':', label=f'Class {i} Train (AUC = {roc_auc_train:.3f})')
        plt.plot(fpr_val, tpr_val, linestyle='--', label=f'Class {i} Val (AUC = {roc_auc_val:.3f})')
        plt.plot(fpr_test, tpr_test, linestyle='-', label=f'Class {i} Test (AUC = {roc_auc_test:.3f})')

    plt.plot([0, 1], [0, 1], 'k--', label='Random')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve - Multi-class')
    plt.legend(loc='best')
    plt.grid(alpha=0.3)
    plt.show()

    # Feature importance visualization for RandomForest in the ensemble
    try:
        rf_model = model.named_estimators_['rf']
        feature_names = X_train.columns
        importances = rf_model.feature_importances_
        indices = np.argsort(importances)[::-1]

        plt.figure(figsize=(12, 6))
        plt.title('Feature Importances from RandomForest Model')
        plt.bar(range(X_train.shape[1]), importances[indices], align='center')
        plt.xticks(range(X_train.shape[1]), [feature_names[i] for i in indices], rotation=90)
        plt.tight_layout()
        plt.show()
    except Exception as e:
        print(f"Couldn't extract feature importances: {e}")

    return {
        'train_acc': train_acc,
        'val_acc': val_acc,
        'test_acc': test_acc,
        'train_report': classification_report(y_train, train_preds, output_dict=True),
        'val_report': classification_report(y_val, val_preds, output_dict=True),
        'test_report': classification_report(y_test, test_preds, output_dict=True)
    }
def plot_learning_curve(estimator, X, y, cv=3, train_sizes=np.linspace(0.1, 1.0, 10)):
    """
    Plots the learning curve of a machine learning model.

    Parameters:
    -----------
    estimator : object
        The machine learning model.
    X : array-like
        Feature data.
    y : array-like
        Target data.
    cv : int or cross-validation generator
        Number of cross-validation folds or a CV generator instance.
    train_sizes : array-like
        Relative or absolute numbers of training examples to use to generate the learning curve.
    """
    plt.figure(figsize=(10, 6))

    # Calculate learning curve values
    train_sizes_abs, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, train_sizes=train_sizes, n_jobs=-1, scoring='accuracy'
    )

    # Calculate mean and standard deviation for training set scores
    train_mean = np.mean(train_scores, axis=1)
    train_std = np.std(train_scores, axis=1)

    # Calculate mean and standard deviation for test set scores
    test_mean = np.mean(test_scores, axis=1)
    test_std = np.std(test_scores, axis=1)

    # Plot learning curve
    plt.plot(train_sizes_abs, train_mean, 'o-', color="r", label="Training score")
    plt.plot(train_sizes_abs, test_mean, 'o-', color="g", label="Cross-validation score")

    # Plot standard deviation bands
    plt.fill_between(train_sizes_abs, train_mean - train_std, train_mean + train_std, alpha=0.1, color="r")
    plt.fill_between(train_sizes_abs, test_mean - test_std, test_mean + test_std, alpha=0.1, color="g")

    # Add labels and title
    plt.xlabel("Training examples")
    plt.ylabel("Accuracy")
    plt.title("Learning Curve")
    plt.legend(loc="best")
    plt.grid(alpha=0.3)

    plt.show()
# =============================================================================
# MAIN EXECUTION
# =============================================================================
def main():
    # 1. Load and explore the dataset
    df = load_dataset("modified_dataset (1).xlsx")
    if df is None:
        return
    exploratory_data_analysis(df)

    # 2. Preprocess the data
    df_preprocessed, le_target = preprocess_data(df)

    # 3. Split the data into training, validation, and test sets
    X_train, X_val, X_test, y_train, y_val, y_test = split_data(df_preprocessed)

    # 4. Train and evaluate models (Stacking Ensemble and NN)
    results = train_and_evaluate_models(X_train, X_val, X_test, y_train, y_val, y_test)
    print("\nEnsemble Model Test Classification Report:")
    print(classification_report(y_test, results['ensemble']['test_preds']))


    """# Plot Learning Curve for the Ensemble Model
    ensemble_model = results['ensemble']['model']
    plot_learning_curve(ensemble_model, X_train, y_train, cv=3)  # Consistent CV

    # 5. Perform hyperparameter tuning for the neural network
    param_grid_nn = {
        'learning_rate': [1e-2, 1e-3, 1e-4],
        'dropout_rate': [0.2, 0.3, 0.4],
        'num_neurons': [32, 64, 128],
        'batch_size': [16, 32, 64],
        'epochs': [50, 100]
    }
    final_nn_model, best_nn_params = hyperparameter_tuning_nn(X_train, y_train, X_val, y_val, param_grid_nn, iterations=10)
"""
    # 4. Get a reduced parameter grid for fast tuning
    param_grid_fast = get_ensemble_param_grid_fast()

    # 5. Perform hyperparameter tuning with fast settings
    best_ensemble, best_ensemble_params = hyperparameter_tuning_ensemble_fast(
        X_train, y_train, X_val, y_val, param_grid_fast, n_iter=5, cv=3
    )

    # 6. Evaluate the tuned ensemble on the validation and test sets
    tuned_val_preds = best_ensemble.predict(X_val)
    tuned_test_preds = best_ensemble.predict(X_test)
    print("\nTuned Ensemble Model Validation Report (Fast Tuning):")
    print(classification_report(y_val, tuned_val_preds))
    print("\nTuned Ensemble Model Test Report (Fast Tuning):")
    print(classification_report(y_test, tuned_test_preds))


    # 9. Save important results and models
    results_dir = "results"
    if not os.path.exists(results_dir):
        os.makedirs(results_dir)

    # Save evaluation results
    evaluation_results = {
        'ensemble': ensemble_evaluation,
        'best_ensemble_params': best_ensemble_params

    }

    try:
        with open(os.path.join(results_dir, "model_evaluation_results.pkl"), "wb") as f:
            pickle.dump(evaluation_results, f)
        print("Evaluation results saved successfully!")
    except Exception as e:
        print(f"Error saving evaluation results: {e}")

    print("\nModel training and evaluation complete!")

if __name__ == "__main__":
    main()
```

```python

# =============================================================================
# SECTION 2: POSTURE AND BACKPACK ANALYSIS USING COMPUTER VISION
# =============================================================================
import cv2
import mediapipe as mp
import numpy as np
from google.colab.patches import cv2_imshow
from google.colab import files
import torch


class BackpackDetector:
    def __init__(self):
        # Load YOLO model
        self.model = torch.hub.load('ultralytics/yolov5', 'yolov5l', pretrained=True)

        # Define objects that could be backpacks or similar items
        self.target_classes = [
            'backpack',
            'handbag',
            'suitcase',
            'bag',
            'sports bag'
        ]

    def detect_backpack(self, image, conf_threshold=0.2):
        """Detect backpack or similar objects in image and return bounding box."""
        results = self.model(image)

        # Get detections as a DataFrame
        detections = results.pandas().xyxy[0]

        # Print all detected objects for debugging
        print("\nDetected objects:")
        for _, det in detections.iterrows():
            print(f"Class: {det['name']}, Confidence: {det['confidence']:.2f}")

        # Filter for target objects with confidence above the threshold
        valid_detections = detections[
            (detections['name'].isin(self.target_classes)) &
            (detections['confidence'] > conf_threshold)
        ].copy()

        # Sort by position (prefer objects closer to back/center)
        if len(valid_detections) > 0:
            # Calculate image center x-coordinate
            center_x = image.shape[1] / 2

            # Add distance from center to detections
            valid_detections['center_dist'] = abs((valid_detections['xmin'] + valid_detections['xmax'])/2 - center_x)

            # Sort by confidence and center distance
            valid_detections = valid_detections.sort_values(
                by=['confidence', 'center_dist'],
                ascending=[False, True]
            )

            best_detection = valid_detections.iloc[0]
            return {
                'x1': int(best_detection['xmin']),
                'y1': int(best_detection['ymin']),
                'x2': int(best_detection['xmax']),
                'y2': int(best_detection['ymax']),
                'confidence': best_detection['confidence'],
                'class': best_detection['name']
            }
        return None

class PostureAnalyzer:
    def __init__(self):
        self.mp_pose = mp.solutions.pose
        self.mp_drawing = mp.solutions.drawing_utils
        self.pose = self.mp_pose.Pose(static_image_mode=True, min_detection_confidence=0.5)
        self.backpack_detector = BackpackDetector()

        # Normal ranges and risk thresholds
        self.normal_ranges = {
            'cervical_curve': (20, 40),
            'thoracic_curve': (20, 45),
            'lumbar_curve': (20, 40),
            'pelvic_tilt': (0, 10),
            'cva': (50, 60),
            'shoulder_alignment': (0, 0),
            'head_tilt': (0, 0)
        }

        self.risk_thresholds = {
            'spine': {
                'low': 5,
                'medium': 10
            },
            'alignment': {
                'low': 2,
                'medium': 5
            }
        }

        self.backpack_thresholds = {
            'optimal_top': 0.3,
            'optimal_bottom': 0.7,
            'medium_bottom': 0.9,
        }



    def get_coords(self, landmark, image_shape):
        """Convert MediaPipe normalized coordinates to pixel coordinates."""
        return np.array([
            int(landmark.x * image_shape[1]),  # x coordinate
            int(landmark.y * image_shape[0])   # y coordinate
        ])

    def calculate_angle(self, a, b, c):
        """Calculate angle between three points (in degrees)."""
        a = np.array(a)
        b = np.array(b)
        c = np.array(c)

        # Create vectors from point b to points a and c
        ba = a - b
        bc = c - b

        # Calculate cosine of angle using dot product
        cosine_angle = np.dot(ba, bc) / (np.linalg.norm(ba) * np.linalg.norm(bc))
        angle = np.arccos(np.clip(cosine_angle, -1.0, 1.0))

        # Convert to degrees
        angle_deg = np.degrees(angle)

        return angle_deg

    def analyze_image(self, image, is_side_view=True):
        """Analyze image and return both pose results and measurements."""
        try:
            image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
            results = self.pose.process(image_rgb)

            if not results.pose_landmarks:
                print("No pose landmarks detected in the image.")
                return None, None

            measurements = self.get_measurements(results.pose_landmarks.landmark, image.shape, is_side_view)
            return results, measurements
        except Exception as e:
            print(f"Error analyzing image: {str(e)}")
            return None, None

    def get_measurements(self, landmarks, image_shape, is_side_view=True):
        """Extract measurements from landmarks."""
        try:
            measurements = {}

            if is_side_view:
                # Side view measurements
                ear = self.get_coords(landmarks[self.mp_pose.PoseLandmark.LEFT_EAR], image_shape)
                shoulder = self.get_coords(landmarks[self.mp_pose.PoseLandmark.LEFT_SHOULDER], image_shape)
                hip = self.get_coords(landmarks[self.mp_pose.PoseLandmark.LEFT_HIP], image_shape)
                knee = self.get_coords(landmarks[self.mp_pose.PoseLandmark.LEFT_KNEE], image_shape)
                ankle = self.get_coords(landmarks[self.mp_pose.PoseLandmark.LEFT_ANKLE], image_shape)

                measurements['cervical_curve'] = self.calculate_angle(ear, shoulder, hip)
                measurements['thoracic_curve'] = self.calculate_angle(shoulder, hip, knee)
                measurements['lumbar_curve'] = self.calculate_angle(hip, knee, ankle)
                measurements['pelvic_tilt'] = self.calculate_angle(
                    hip,
                    np.array([hip[0] + 100, hip[1]]),
                    shoulder
                )
                measurements['cva'] = self.calculate_angle(
                    ear,
                    shoulder,
                    np.array([shoulder[0] + 100, shoulder[1]])
                )
            else:
                # Front view measurements
                left_shoulder = self.get_coords(landmarks[self.mp_pose.PoseLandmark.LEFT_SHOULDER], image_shape)
                right_shoulder = self.get_coords(landmarks[self.mp_pose.PoseLandmark.RIGHT_SHOULDER], image_shape)
                left_ear = self.get_coords(landmarks[self.mp_pose.PoseLandmark.LEFT_EAR], image_shape)
                right_ear = self.get_coords(landmarks[self.mp_pose.PoseLandmark.RIGHT_EAR], image_shape)

                measurements['shoulder_alignment'] = abs(left_shoulder[1] - right_shoulder[1])
                measurements['head_tilt'] = abs(left_ear[1] - right_ear[1])

            return measurements
        except Exception as e:
            print(f"Error getting measurements: {str(e)}")
            return None

    def assess_deviation(self, neutral_measurements, loaded_measurements):
        """Calculate deviations between neutral and loaded postures."""
        deviations = {}
        risk_assessments = {}

        for measure in neutral_measurements:
            deviation = abs(neutral_measurements[measure] - loaded_measurements[measure])
            deviations[measure] = deviation

            # Determine risk category
            category = 'alignment' if measure in ['shoulder_alignment', 'head_tilt'] else 'spine'
            thresholds = self.risk_thresholds[category]

            if deviation < thresholds['low']:
                risk_level = 'Low Risk'
            elif deviation < thresholds['medium']:
                risk_level = 'Medium Risk'
            else:
                risk_level = 'High Risk'

            risk_assessments[measure] = (risk_level, deviation)

        return deviations, risk_assessments

    def generate_feedback(self, risk_assessments):
        """Generate feedback and suggestions based on risk assessments."""
        feedback = []

        for measure, (risk_level, deviation) in risk_assessments.items():
            if risk_level == 'High Risk':
                if measure == 'cervical_curve':
                    feedback.append(f"High risk in neck posture (deviation: {deviation:.1f}°). "
                                 "Suggestions:\n"
                                 "- Adjust backpack straps to distribute weight evenly\n"
                                 "- Consider reducing load weight\n"
                                 "- Practice neck strengthening exercises")

                elif measure == 'thoracic_curve':
                    feedback.append(f"High risk in upper back posture (deviation: {deviation:.1f}°). "
                                 "Suggestions:\n"
                                 "- Use both shoulder straps\n"
                                 "- Tighten straps to keep load close to body\n"
                                 "- Consider a backpack with chest strap")

                elif measure == 'lumbar_curve':
                    feedback.append(f"High risk in lower back posture (deviation: {deviation:.1f}°). "
                                 "Suggestions:\n"
                                 "- Use a backpack with padded back support\n"
                                 "- Distribute weight evenly\n"
                                 "- Consider hip belt for weight distribution")

                elif measure == 'shoulder_alignment':
                    feedback.append(f"High risk in shoulder alignment (deviation: {deviation:.1f}°). "
                                 "Suggestions:\n"
                                 "- Adjust strap lengths to be equal\n"
                                 "- Use both shoulder straps\n"
                                 "- Redistribute weight in backpack")

                elif measure == 'head_tilt':
                    feedback.append(f"High risk in head alignment (deviation: {deviation:.1f}°). "
                                 "Suggestions:\n"
                                 "- Check for uneven strap tension\n"
                                 "- Practice neck alignment exercises\n"
                                 "- Consider lighter load")

        if not feedback:
            feedback.append("No high-risk deviations detected. Continue maintaining good posture and regular monitoring.")

        return "\n\n".join(feedback)

    def generate_feedback(self, risk_assessments):
        """Generate feedback and suggestions based on risk assessments."""
        feedback = []

        for measure, (risk_level, deviation) in risk_assessments.items():
            if risk_level == 'High Risk':
                if measure == 'cervical_curve':
                    feedback.append(f"High risk in neck posture (deviation: {deviation:.1f}°). "
                                 "Suggestions:\n"
                                 "- Adjust backpack straps to distribute weight evenly\n"
                                 "- Consider reducing load weight\n"
                                 "- Practice neck strengthening exercises")

                elif measure == 'thoracic_curve':
                    feedback.append(f"High risk in upper back posture (deviation: {deviation:.1f}°). "
                                 "Suggestions:\n"
                                 "- Use both shoulder straps\n"
                                 "- Tighten straps to keep load close to body\n"
                                 "- Consider a backpack with chest strap")

                elif measure == 'lumbar_curve':
                    feedback.append(f"High risk in lower back posture (deviation: {deviation:.1f}°). "
                                 "Suggestions:\n"
                                 "- Use a backpack with padded back support\n"
                                 "- Distribute weight evenly\n"
                                 "- Consider hip belt for weight distribution")

                elif measure == 'shoulder_alignment':
                    feedback.append(f"High risk in shoulder alignment (deviation: {deviation:.1f}°). "
                                 "Suggestions:\n"
                                 "- Adjust strap lengths to be equal\n"
                                 "- Use both shoulder straps\n"
                                 "- Redistribute weight in backpack")

                elif measure == 'head_tilt':
                    feedback.append(f"High risk in head alignment (deviation: {deviation:.1f}°). "
                                 "Suggestions:\n"
                                 "- Check for uneven strap tension\n"
                                 "- Practice neck alignment exercises\n"
                                 "- Consider lighter load")

        if not feedback:
            feedback.append("No high-risk deviations detected. Continue maintaining good posture and regular monitoring.")

        return "\n\n".join(feedback)

    def analyze_backpack_position(self, image, pose_results):
        """Analyze backpack position and assess risk level."""
        if not pose_results or not pose_results.pose_landmarks:
            return None, "No pose landmarks detected."

        backpack = self.backpack_detector.detect_backpack(image)
        if not backpack:
            return None, "No backpack or similar object detected in the image."

        print(f"\nDetected object: {backpack['class']} with {backpack['confidence']:.2f} confidence")

        landmarks = pose_results.pose_landmarks.landmark

        # Get key body landmarks
        shoulder_y = min(
            landmarks[self.mp_pose.PoseLandmark.LEFT_SHOULDER].y,
            landmarks[self.mp_pose.PoseLandmark.RIGHT_SHOULDER].y
        ) * image.shape[0]

        hip_y = min(
            landmarks[self.mp_pose.PoseLandmark.LEFT_HIP].y,
            landmarks[self.mp_pose.PoseLandmark.RIGHT_HIP].y
        ) * image.shape[0]

        # Calculate relative position of backpack
        backpack_top = backpack['y1']
        backpack_bottom = backpack['y2']

        # Calculate relative positions
        torso_length = hip_y - shoulder_y
        if torso_length <= 0:
            return None, "Could not determine torso length accurately."

        backpack_top_relative = (backpack_top - shoulder_y) / torso_length
        backpack_bottom_relative = (backpack_bottom - shoulder_y) / torso_length

        # Assess risk level
        if (backpack_top_relative >= self.backpack_thresholds['optimal_top'] and
            backpack_bottom_relative <= self.backpack_thresholds['optimal_bottom']):
            risk_level = 'Low Risk'
            message = (f"Detected {backpack['class']} position is optimal - between shoulder blades and lower back. "
                      "This helps maintain natural spine alignment.")

        elif backpack_bottom_relative <= self.backpack_thresholds['medium_bottom']:
            risk_level = 'Medium Risk'
            message = (f"Detected {backpack['class']} sits slightly low on the back. Consider adjusting straps "
                      "to raise the position and maintain better posture.")

        else:
            risk_level = 'High Risk'
            message = (f"Detected {backpack['class']} position is too low, causing potential forward lean and "
                      "increased stress on lower back. Adjust straps to raise "
                      "position significantly.")

        return {
            'risk_level': risk_level,
            'top_position': backpack_top_relative,
            'bottom_position': backpack_bottom_relative,
            'bbox': backpack
        }, message

    def visualize_backpack_analysis(self, image, backpack_analysis):
        """Visualize backpack position and risk level on image."""
        if not backpack_analysis:
            return image

        result = image.copy()
        bbox = backpack_analysis['bbox']
        risk_level = backpack_analysis['risk_level']

        # Color based on risk level
        color = {
            'Low Risk': (0, 255, 0),
            'Medium Risk': (0, 255, 255),
            'High Risk': (0, 0, 255)
        }[risk_level]

        # Draw backpack bounding box
        cv2.rectangle(result,
                     (bbox['x1'], bbox['y1']),
                     (bbox['x2'], bbox['y2']),
                     color, 2)

        # Add detected class and risk level text
        cv2.putText(result,
                   f"Detected: {bbox['class']} ({risk_level})",
                   (10, image.shape[0] - 20),
                   cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)

        return result
def main():
  try:
    analyzer = PostureAnalyzer()

    # 1. Collect and analyze neutral posture images
    print("Upload NEUTRAL SIDE VIEW image:")
    uploaded = files.upload()
    neutral_side = cv2.imread(list(uploaded.keys())[0])
    neutral_side_results, neutral_side_measurements = analyzer.analyze_image(neutral_side, is_side_view=True)

    print("\nUpload NEUTRAL FRONT VIEW image:")
    uploaded = files.upload()
    neutral_front = cv2.imread(list(uploaded.keys())[0])
    neutral_front_results, neutral_front_measurements = analyzer.analyze_image(neutral_front, is_side_view=False)

    # 2. Collect and analyze loaded posture images
    print("\nUpload BACKPACK LOADED SIDE VIEW image:")
    uploaded = files.upload()
    loaded_side = cv2.imread(list(uploaded.keys())[0])
    loaded_side_results, loaded_side_measurements = analyzer.analyze_image(loaded_side, is_side_view=True)

    print("\nUpload BACKPACK LOADED FRONT VIEW image:")
    uploaded = files.upload()
    loaded_front = cv2.imread(list(uploaded.keys())[0])
    loaded_front_results, loaded_front_measurements = analyzer.analyze_image(loaded_front, is_side_view=False)

    if all([neutral_side_measurements, neutral_front_measurements,
            loaded_side_measurements, loaded_front_measurements]):

        # 4. Calculate deviations and assess risks
        side_deviations, side_risks = analyzer.assess_deviation(
            neutral_side_measurements, loaded_side_measurements)
        front_deviations, front_risks = analyzer.assess_deviation(
            neutral_front_measurements, loaded_front_measurements)

        # 5. Display results
        print("\nSIDE VIEW MEASUREMENTS:")
        print("----------------------")
        for measure in side_deviations:
            risk_level, deviation = side_risks[measure]
            print(f"{measure}:")
            print(f"  Neutral: {neutral_side_measurements[measure]:.1f}°")
            print(f"  Loaded:  {loaded_side_measurements[measure]:.1f}°")
            print(f"  Deviation: {deviation:.1f}° ({risk_level})")

        print("\nFRONT VIEW MEASUREMENTS:")
        print("-----------------------")
        for measure in front_deviations:
            risk_level, deviation = front_risks[measure]
            print(f"{measure}:")
            print(f"  Neutral: {neutral_front_measurements[measure]:.1f}°")
            print(f"  Loaded:  {loaded_front_measurements[measure]:.1f}°")
            print(f"  Deviation: {deviation:.1f}° ({risk_level})")

        # 6. Generate and display feedback
        print("\nFEEDBACK AND SUGGESTIONS:")
        print("-----------------------")
        all_risks = {**side_risks, **front_risks}
        feedback = analyzer.generate_feedback(all_risks)
        print(feedback)

    else:
        print("Error: Could not detect landmarks in one or more images. "
              "Please ensure full body visibility in all photos.")


               # 3. Analyze backpack position
    print("\nANALYZING BACKPACK POSITION:")
    print("---------------------------")
    if loaded_side_results and loaded_side_measurements:  # Check both results
            backpack_analysis, backpack_message = analyzer.analyze_backpack_position(
                loaded_side, loaded_side_results)  # Pass pose_results instead of landmarks

            if backpack_analysis:
                print(f"Risk Level: {backpack_analysis['risk_level']}")
                print(f"Analysis: {backpack_message}")

                # Visualize backpack position
                annotated_image = analyzer.visualize_backpack_analysis(
                    loaded_side, backpack_analysis)
                cv2_imshow(annotated_image)

                # Add backpack position recommendations to feedback
                if backpack_analysis['risk_level'] != 'Low Risk':
                    print("\nBACKPACK POSITION RECOMMENDATIONS:")
                    print("--------------------------------")
                    print("1. Adjust shoulder straps to position backpack higher on back")
                    print("2. Use chest strap if available to secure position")
                    print("3. Pack heavier items closer to back and higher in the backpack")
                    if backpack_analysis['risk_level'] == 'High Risk':
                        print("4. Consider using a different backpack with better support features")
                        print("5. Reduce load weight if possible")
            else:
                print(f"Backpack analysis failed: {backpack_message}")
    else:
            print("Could not analyze backpack position. Please ensure both pose landmarks and backpack are visible in side view photo.")

  except Exception as e:
        print(f"An error occurred: {str(e)}")
        import traceback
        traceback.print_exc()

if __name__ == "__main__":
    main()

import cv2
import mediapipe as mp
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle  # Use pickle to load the ensemble model

import tensorflow as tf  # Only needed if you want to compare with NN model



# We assume that the PostureAnalyzer class (with BackpackDetector, get_coords, etc.) is defined as before.

class PostureMSDAnalyzer:

    def __init__(self, model_path="/content/best_ensemble_model_fast.pkl"):

        # Initialize the PostureAnalyzer (from your computer vision section)

        self.posture_analyzer = PostureAnalyzer()

        # Load the ensemble model using pickle

        with open(model_path, 'rb') as f:

            self.ml_model = pickle.load(f)

        # Risk level mapping (index to label)
        self.risk_levels = ['Low', 'Medium', 'High']

    def get_user_inputs(self):
        """Collect manual inputs from user."""
        inputs = {}
        inputs['Age'] = float(input("Enter age: "))

        # Convert text input to numeric for gender
        gender = input("Enter gender (Male/Female): ").strip().lower()
        inputs['Gender'] = 1 if gender == 'male' else 0

        inputs['Height'] = float(input("Enter height in cm: "))
        inputs['Weight'] = float(input("Enter weight in kg: "))
        inputs['Pain_Score'] = float(input("Enter pain score (0-10): "))
        inputs['Backpack_Weight'] = float(input("Enter backpack weight in kg: "))
        inputs['Backpack_Height'] = float(input("Enter backpack height in cm: "))
        inputs['Daily_Load_Duration'] = float(input("Enter daily load duration in minutes: "))
        return inputs

    def prepare_ml_features(self, user_inputs, neutral_measurements, loaded_measurements, backpack_analysis):
        features = {}
        # Base user inputs
        features.update(user_inputs)

        # Instead of a single numeric encoding, perform one-hot encoding for Backpack_Position
        if backpack_analysis and 'risk_level' in backpack_analysis:
            if backpack_analysis['risk_level'] == 'Low Risk':
                backpack_pos = 'between the shoulder blades and lower back'
            elif backpack_analysis['risk_level'] == 'Medium Risk':
                backpack_pos = 'slightly lower on the back'
            else:
                backpack_pos = 'on the buttocks or significantly low'
        else:
            backpack_pos = 'slightly lower on the back'  # Default

        # One-hot encoding for backpack position
        backpack_pos_mapping = {
            'between the shoulder blades and lower back': [1, 0, 0],
            'slightly lower on the back': [0, 1, 0],
            'on the buttocks or significantly low': [0, 0, 1]
        }
        one_hot_values = backpack_pos_mapping.get(backpack_pos, [0, 1, 0])
        features['Backpack_Pos_1'] = one_hot_values[0]
        features['Backpack_Pos_2'] = one_hot_values[1]
        features['Backpack_Pos_3'] = one_hot_values[2]

        # Add posture measurement features from neutral images
        features['Neutral_Cervical_Curve'] = neutral_measurements.get('cervical_curve', 0)
        features['Neutral_Thoracic_Curve'] = neutral_measurements.get('thoracic_curve', 0)
        features['Neutral_Lumbar_Curve'] = neutral_measurements.get('lumbar_curve', 0)
        features['Neutral_Pelvic_Tilt'] = neutral_measurements.get('pelvic_tilt', 0)
        features['Neutral_CVA'] = neutral_measurements.get('cva', 0)
        features['Neutral_Shoulder_Alignment'] = neutral_measurements.get('shoulder_alignment', 0)
        features['Neutral_Head_Tilt'] = neutral_measurements.get('head_tilt', 0)

        # Calculate deviations between loaded and neutral measurements
        features['Cervical_Deviation'] = abs(loaded_measurements.get('cervical_curve', 0) - features['Neutral_Cervical_Curve'])
        features['Thoracic_Deviation'] = abs(loaded_measurements.get('thoracic_curve', 0) - features['Neutral_Thoracic_Curve'])
        features['Lumbar_Deviation'] = abs(loaded_measurements.get('lumbar_curve', 0) - features['Neutral_Lumbar_Curve'])
        features['Pelvic_Tilt_Deviation'] = abs(loaded_measurements.get('pelvic_tilt', 0) - features['Neutral_Pelvic_Tilt'])
        features['CVA_Deviation'] = abs(loaded_measurements.get('cva', 0) - features['Neutral_CVA'])
        features['Shoulder_Deviation'] = abs(loaded_measurements.get('shoulder_alignment', 0) - features['Neutral_Shoulder_Alignment'])
        features['Head_Tilt_Deviation'] = abs(loaded_measurements.get('head_tilt', 0) - features['Neutral_Head_Tilt'])

        # Add interaction/polynomial features as done in training
        features['Weight_BackpackWeight'] = user_inputs['Weight'] * user_inputs['Backpack_Weight']
        features['Age_PainScore'] = user_inputs['Age'] * user_inputs['Pain_Score']
        features['Pain_Score_Squared'] = user_inputs['Pain_Score'] ** 2
        features['Backpack_Weight_Cubed'] = user_inputs['Backpack_Weight'] ** 3
        features['Age_Squared'] = user_inputs['Age'] ** 2

        # Assemble final features in the same order as training
        final_feature_order = [
            'Age', 'Gender', 'Height', 'Weight', 'Pain_Score', 'Backpack_Weight',
            'Backpack_Height', 'Daily_Load_Duration',
            'Backpack_Pos_1', 'Backpack_Pos_2', 'Backpack_Pos_3',
            'Neutral_Cervical_Curve', 'Neutral_Thoracic_Curve', 'Neutral_Lumbar_Curve',
            'Neutral_Pelvic_Tilt', 'Neutral_CVA', 'Neutral_Shoulder_Alignment', 'Neutral_Head_Tilt',
            'Cervical_Deviation', 'Thoracic_Deviation', 'Lumbar_Deviation', 'Pelvic_Tilt_Deviation',
            'CVA_Deviation', 'Shoulder_Deviation', 'Head_Tilt_Deviation',
            'Weight_BackpackWeight', 'Age_PainScore', 'Pain_Score_Squared',
            'Backpack_Weight_Cubed', 'Age_Squared'
        ]

        # Create the input array for prediction
        numeric_features = [features[name] for name in final_feature_order]
        input_array = np.array([numeric_features])
        return input_array


    def encode_backpack_position(self, position):
        """Convert backpack position to numeric encoding."""
        position_mapping = {
            'between the shoulder blades and lower back': 0,
            'slightly lower on the back': 1,
            'on the buttocks or significantly low': 2
        }
        return position_mapping.get(position, 1)  # Default to middle position if unknown

    def predict_msd_risk(self, features):
        """Predict MSD risk using the ML model."""
        # If 'features' is already a NumPy array, assume it's in the correct order.
        if isinstance(features, np.ndarray):
            input_array = features
        else:
            # Define the feature order matching your training data
            feature_names = [
                'Age', 'Gender', 'Height', 'Weight', 'Pain_Score', 'Backpack_Weight',
                'Backpack_Height', 'Daily_Load_Duration',
                'Backpack_Pos_1', 'Backpack_Pos_2', 'Backpack_Pos_3',
                'Neutral_Cervical_Curve', 'Neutral_Thoracic_Curve', 'Neutral_Lumbar_Curve',
                'Neutral_Pelvic_Tilt', 'Neutral_CVA', 'Neutral_Shoulder_Alignment', 'Neutral_Head_Tilt',
                'Cervical_Deviation', 'Thoracic_Deviation', 'Lumbar_Deviation', 'Pelvic_Tilt_Deviation',
                'CVA_Deviation', 'Shoulder_Deviation', 'Head_Tilt_Deviation',
                'Weight_BackpackWeight', 'Age_PainScore', 'Pain_Score_Squared',
                'Backpack_Weight_Cubed', 'Age_Squared'
            ]
            numeric_features = [features[name] for name in feature_names]
            input_array = np.array([numeric_features])

        # Get prediction probabilities from the model
        prediction_probs = self.ml_model.predict_proba(input_array)[0]

        # Determine risk level and calculate confidence
        risk_level = self.risk_levels[np.argmax(prediction_probs)]
        confidence = prediction_probs[np.argmax(prediction_probs)] * 100

        return risk_level, confidence


    def analyze_and_predict(self):
        """Main function to run complete analysis and prediction."""
        try:
            # 1. Get user inputs
            print("\nCollecting user information...")
            user_inputs = self.get_user_inputs()

            # 2. Analyze posture images
            print("\nAnalyzing neutral posture...")
            print("Upload NEUTRAL SIDE VIEW image:")
            neutral_side = cv2.imread(input("Enter path to neutral side view image: "))
            neutral_side_results, neutral_side_measurements = (
                self.posture_analyzer.analyze_image(neutral_side, is_side_view=True))

            print("\nUpload NEUTRAL FRONT VIEW image:")
            neutral_front = cv2.imread(input("Enter path to neutral front view image: "))
            neutral_front_results, neutral_front_measurements = (
                self.posture_analyzer.analyze_image(neutral_front, is_side_view=False))

            print("\nAnalyzing loaded posture...")
            print("Upload LOADED SIDE VIEW image:")
            loaded_side = cv2.imread(input("Enter path to loaded side view image: "))
            loaded_side_results, loaded_side_measurements = (
                self.posture_analyzer.analyze_image(loaded_side, is_side_view=True))

            print("\nUpload LOADED FRONT VIEW image:")
            loaded_front = cv2.imread(input("Enter path to loaded front view image: "))
            loaded_front_results, loaded_front_measurements = (
                self.posture_analyzer.analyze_image(loaded_front, is_side_view=False))

            print("\nAnalyzing backpack position:")

            # 3. Analyze backpack position
            backpack_analysis, backpack_message = (
                self.posture_analyzer.analyze_backpack_position(loaded_side, loaded_side_results))

            # 4. Combine measurements
            neutral_measurements = {**neutral_side_measurements, **neutral_front_measurements}
            loaded_measurements = {**loaded_side_measurements, **loaded_front_measurements}

            # 5. Prepare features and predict
            features = self.prepare_ml_features(
                user_inputs, neutral_measurements, loaded_measurements, backpack_analysis)
            risk_level, confidence = self.predict_msd_risk(features)

            # 6. Display results
            print("\nANALYSIS RESULTS:")
            print("-----------------")
            print(f"MSD Risk Level: {risk_level}")
            print(f"Confidence: {confidence:.1f}%")
            print(f"\nBackpack Position Analysis: {backpack_message}")

            # Calculate and display posture deviations
            side_deviations, side_risks = self.posture_analyzer.assess_deviation(
                neutral_side_measurements, loaded_side_measurements)
            front_deviations, front_risks = self.posture_analyzer.assess_deviation(
                neutral_front_measurements, loaded_front_measurements)

            print("\nPosture Analysis:")
            all_risks = {**side_risks, **front_risks}
            feedback = self.posture_analyzer.generate_feedback(all_risks)
            print(feedback)

            return risk_level, backpack_analysis, all_risks

        except Exception as e:
            print(f"Error during analysis: {str(e)}")
            import traceback
            traceback.print_exc()
            return None, None, None

def main():
    # Initialize analyzer
    analyzer = PostureMSDAnalyzer(model_path="best_ensemble_model_fast.pkl")

    # Run analysis
    risk_level, backpack_analysis, posture_risks = analyzer.analyze_and_predict()

    if risk_level is not None:
        print("\nAnalysis completed successfully!")
        print(f"Final MSD Risk Assessment: {risk_level}")
        print("Use these results to make informed decisions about backpack usage and posture.")

if __name__ == "__main__":
    main()
```
